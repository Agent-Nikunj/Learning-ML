```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples = 1000, centers = 3, n_features = 2, shuffle = True, random_state = 31)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
x
```

```python
kmeans = KMeans(init = "random", n_clusters=5, random_state=0, n_init= 5).fit(x, y=None, sample_weight=None)
kmeans.labels_
```

```python
result = kmeans.predict(x)
```

```python
centers = kmeans.cluster_centers_
```

```python
labels = np.unique(result)
```

```python
for i in labels:
    plt.scatter(x[:,0],x[:,1], c = kmeans.labels_)

plt.scatter(centers[:,0],centers[:,1],marker = '^', color = 'r', zorder = 5)
plt.show()
```
